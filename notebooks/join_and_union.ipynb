```python
import pandas as pd
from datetime import datetime
```

```python
# Customer data
customer_data = {
    'id': [1, 2, 3, 4, 5],
    'name': ['John Smith', 'Jane Doe', 'Bob Wilson', 'Mary Johnson', 'Steve Brown'],
    'region': ['North', 'South', 'East', 'North', 'South'],
    'membership_level': ['Gold', 'Silver', 'Bronze', 'Gold', 'Silver']
}

# Order data 
order_data = {
    'order_id': [101, 102, 103, 104, 105, 106, 107],
    'customer_id': [1, 2, 3, 4, 5, 1, None],
    'product': ['A', 'B', 'C', 'A', None, 'A', 'C'],
    'quantity': [2, 1, 5, 3, 2, None, 1],
    'price': [10.5, 20.0, 7.5, 10.5, 20.0, 10.5, None],
    'purchase_date': [
        datetime(2025, 4, 1),
        datetime(2025, 4, 2), 
        datetime(2025, 4, 2),
        datetime(2025, 4, 3),
        datetime(2025, 4, 4),
        None,
        datetime(2025, 4, 5)
    ],
    'ship_date': [
        datetime(2025, 4, 2),
        datetime(2025, 4, 3),
        datetime(2025, 4, 3), 
        datetime(2025, 4, 4),
        datetime(2025, 4, 7),
        datetime(2025, 4, 6),
        None
    ],
    'region': ['North', 'South', 'East', 'North', 'South', 'North', 'East']
}

# Create DataFrames
customers_df = pd.DataFrame(customer_data)
orders_df = pd.DataFrame(order_data)
```

# 1. Merge two DataFrames on different keys

```python
orders_df.merge(customers_df, left_on='customer_id', right_on='id', how='inner')
```

# 2. Merge using `how='left'`, `how='right'`, or `how='outer'`

```python
orders_df.merge(customers_df, on='region', how='left')
```

```python
orders_df.merge(customers_df, on='region', how='right')
```

```python
orders_df.merge(customers_df, on='region', how='outer')
```

# 3. Union two DataFrames (like SQL `UNION ALL`)

```python
pd.concat([customers_df, orders_df], ignore_index=True)
```


# 4. Union with deduplication (like SQL `UNION`)

```python
pd.concat([customers_df, orders_df], ignore_index=True).drop_duplicates()
```
